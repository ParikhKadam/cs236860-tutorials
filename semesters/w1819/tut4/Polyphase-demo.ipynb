## Polyphase using Scipy

```python
from scipy import signal 

x = np.linspace(0, 10, 10, endpoint=False)
y = np.cos(-x**2/6.0)


f_fft = signal.resample(y, 100)
f_poly = signal.resample_poly(y, up=100, down=10)#window=)
xnew = np.linspace(0, 10, 100, endpoint=False)

# print(len(f_fft))
# print(len(f_poly))

plt.figure()
plt.plot(x,y,'ko-')
plt.plot(xnew, f_fft,'bo')
plt.plot(xnew, f_poly,'ro')
plt.legend(['data', 'resamp-fft', 'remsamp-poly'], loc='best')
plt.show()
```
